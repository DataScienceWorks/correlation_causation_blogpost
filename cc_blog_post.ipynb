## Understanding Correlation and Causality by example

This notebook contains code that is used to generate the graph, used in the blog post "Correlation and Causality".

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def prices(min_price, max_price, count=12, rounding=2):
    lst = np.random.uniform(min_price, max_price, count)
    return np.round(lst, rounding)

petrol = prices(60.0, 75.0)
print('Petrol Prices per Litre : ', petrol)
diesel = petrol - prices(20.22,25.55)
print('Diesel Prices per Litre : ', diesel)
crude_oil = diesel - prices(5,10)
print('Crude Oil Prices per Litre : ', crude_oil)
water = [25]*3 + [30]*3 + [25]*6
print('Water Prices per Litre : ', water)
coke = [45]*6 + [30]*6
print('Coke Prices per Litre : ', coke)

df = pd.DataFrame({
    'petrol' : petrol,
    'diesel' : diesel,
    'crude_oil' : crude_oil,
    'water' : water,
    'coke' : coke
})
df.index = df.index + 1

# Show Plot/Graph
df.plot(kind='line', figsize=(8,5), rot=45, color=['orange', 'brown', 'black', 'blue', 'red'])
plt.title('Price graph of goods, for the year XXXX in India')
plt.xlabel('Months')
xticks=['Jan','Feb','Mar','April','May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(df.index.values, xticks) # Using months instead of numbers for xticks
plt.ylabel('Price in INR')
plt.gca().set_ylim(0.5,max(petrol)+10) # Adjusting y-axis
plt.legend(loc='lower center', bbox_to_anchor=(0.5, 0.01), ncol=5) # Adjust legend placement
plt.show()
```
